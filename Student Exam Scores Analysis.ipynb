### 3. Which ethnic group has the highest and lowest average scores in each subject (math, reading, writing)?

```python
import pandas as pd


# Step 2: Group by EthnicGroup and calculate the overall average score
average_scores_by_group = df.groupby(
    'EthnicGroup')['avg_score'].mean().round(2).reset_index()

# Rename columns for clarity
average_scores_by_group.columns = ['EthnicGroup', 'Average Score']

# Display the results
print(average_scores_by_group)
```

```python
import pandas as pd

average_scores_by_group = df.groupby(
    'EthnicGroup')[['MathScore', 'ReadingScore', 'WritingScore']].mean()

results = {}

for subject in ['MathScore', 'ReadingScore', 'WritingScore']:
    highest_group = average_scores_by_group[subject].idxmax()
    highest_score = average_scores_by_group[subject].max()
    lowest_group = average_scores_by_group[subject].idxmin()
    lowest_score = average_scores_by_group[subject].min()

    results[subject] = {
        'Highest': {'Group': highest_group, 'Score': highest_score},
        'Lowest': {'Group': lowest_group, 'Score': lowest_score}
    }

for subject, result in results.items():
    print(f"{subject}:")
    print(
        f"Highest: {result['Highest']['Group']} ({result['Highest']['Score']:.2f})")
    print(
        f"Lowest: {result['Lowest']['Group']} ({result['Lowest']['Score']:.2f})")
    print()
```

### 13. What is the overall pass rate (percentage of students scoring above 50%) in each subject?  

```python
import pandas as pd

pass_threshold = 50

pass_rates = {
    'MathScore': (df['MathScore'] >= pass_threshold).mean() * 100,
    'ReadingScore': (df['ReadingScore'] >= pass_threshold).mean() * 100,
    'WritingScore': (df['WritingScore'] >= pass_threshold).mean() * 100
}

# Display the results
for subject, rate in pass_rates.items():
    print(f"{subject}: {rate:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert pass rates to a DataFrame for visualization
pass_rates_df = pd.DataFrame(list(pass_rates.items()), columns=[
                             'Subject', 'PassRate'])

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=pass_rates_df, x='Subject', y='PassRate', palette="Set2")
plt.title("Overall Pass Rate (Percentage of Students Scoring Above 50%)")
plt.ylabel("Pass Rate (%)")
plt.xlabel("Subject")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

### 14. What percentage of students scored above 80% in all three subjects?  

```python
threshold = 80

# Create a boolean mask for students scoring above 80% in all three subjects
above_80_all_subjects = (
    (df['MathScore'] >= threshold) &
    (df['ReadingScore'] >= threshold) &
    (df['WritingScore'] >= threshold)
)

# Calculate the percentage of students meeting the criterion
percentage_above_80_all_subjects = above_80_all_subjects.mean() * 100

# Display the result
print(f"Percentage of students scoring above 80% in all three subjects: {percentage_above_80_all_subjects:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Count students above and below the threshold
students_above_80 = above_80_all_subjects.sum()
students_below_80 = len(df) - students_above_80

# Create a pie chart
labels = ['Above 80% in All Subjects', 'Below 80% in Any Subject']
sizes = [students_above_80, students_below_80]
colors = ['#4CAF50', '#FFC107']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Proportion of Students Scoring Above 80% in All Three Subjects")
plt.tight_layout()
plt.show()
```
